```python
import pandas as pd

a = [1, 2, 3]

myvar = pd.Series(a)

print(myvar)
```

```python
import pandas as pd

sites = {1: "Google", 2: "Runoob", 3: "Wiki"}

myvar = pd.Series(sites, index = [1, 2])

print(myvar)
```

```python
import pandas as pd

data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

# 数据载入到 DataFrame 对象
df = pd.DataFrame(data)

# 返回第一行和第二行
df
```

```python
import pandas as pd
val_accs_mag_bert = [0.1,0.3,0.8,0.9]
val_accs_cmgcni = [0.2,0.5,0.88,0.98]
data = {
    'mag_bert':val_accs_mag_bert,
    'cmgcni':val_accs_cmgcni
}
df = pd.DataFrame(data)
df.to_csv('7.csv',index=False)
df
```

```python
import pandas as pd
%matplotlib inline
df = pd.read_csv('7.csv')
import matplotlib
import matplotlib.pyplot as plt

fig = df.plot(title = 'val_acc',figsize=(5, 4), fontsize=26).get_figure()
fig.savefig('7.jpg')
plt.show()
```
